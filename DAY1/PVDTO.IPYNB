## 1) STRING (Text)

```python
# String is text data (words, sentences, etc.)

name = "Nirmal saud"
message = 'Hello'
long_text = """
This is a long text
That spans multiple lines
Also called a multi-line string
"""

print(name)
print(type(name))  # <class 'str'>

# Accessing individual characters using index
text = "Hello"
print(text[0])    # H (first character, starts at 0)
print(text[1])    # e (second character)
print(text[-1])   # o (last character)
print(text[-2])   # l (second last)

# Length of string
text = "Python"
print(len(text))  # 6

# String methods (built-in functions)
text = "hello world"
print(text.upper())           # HELLO WORLD
print(text.capitalize())      # Hello world
print(text.title())           # Hello World
print(text.replace("world", "python"))  # hello python
print(text.split(" "))        # ['hello', 'world']

# String concatenation (joining)
first = "Nirmal"
last = "Saud"
full = first + " " + last
print(full)  # Raj Sharma

# String formatting
name = "Nirmal"
age = 18
print(f"My name is {name} and I am {age} years old")  # f-string (best way)
print("My name is {} and I am {} years old".format(name, age))  # format method
print("My name is %s and I am %d years old" % (name, age))  # % method (old)
```

## 4 BOOLEAN (True/False)

```python
# Boolean has only two values: True or False

is_student = True
is_teacher = False

print(is_student)
print(type(is_student))  # <class 'bool'>

# Boolean from comparisons
age = 20
print(age > 18)    # True
print(age < 18)    # False
print(age == 20)   # True
print(age != 20)   # False

# Logical operations
a = True
b = False
print(a and b)     # False
print(a or b)      # True
print(not a)       # False

# Boolean in if statements
password = "hello123"
correct = "hello123"
if password == correct:
    print("Login successful")
```

## 5 LIST (Ordered Collection)

```python
# List is a collection of items (in order)
# List can contain different data types
# Lists are mutable (can be changed)

fruits = ["Apple", "Banana", "Orange", "Mango"]
numbers = [10, 20, 30, 40, 50]
mixed = [10, "Nirmal", 99.99, True, None]

print(fruits)
print(type(fruits))  # <class 'list'>

# Accessing items by index
print(fruits[0])     # Apple (first item, index starts at 0)
print(fruits[1])     # Banana
print(fruits[-1])    # Mango (last item)
print(fruits[-2])    # Orange (second last)

# Length of list
print(len(fruits))   # 4

# Adding items
fruits.append("Grapes")  # Add at the end
print(fruits)

# Insert at specific position
fruits.insert(1, "Papaya")  # Insert at index 1
print(fruits)

# Removing items
fruits.remove("Banana")  # Remove by value
print(fruits)

del fruits[0]  # Remove by index
print(fruits)

# Changing items
fruits[0] = "Pineapple"
print(fruits)

# List slicing (getting a portion)
numbers = [10, 20, 30, 40, 50]
print(numbers[1:3])    # [20, 30] (from index 1 to 2)
print(numbers[:3])     # [10, 20, 30] (first 3)
print(numbers[2:])     # [30, 40, 50] (from index 2 onwards)

# Loop through list
for fruit in fruits:
    print(fruit)

# Useful list methods
numbers = [3, 1, 4, 1, 5]
print(numbers.count(1))    # 2 (count how many 1s)
print(numbers.index(4))    # 2 (index of 4)
numbers.sort()
print(numbers)             # [1, 1, 3, 4, 5]
numbers.reverse()
print(numbers)             # [5, 4, 3, 1, 1]
```

## 6 TUPLE (Immutable Ordered Collection)

```python
# Tuple is like a list but CANNOT be changed
# Once created, you cannot add, remove, or change items

coordinates = (10, 20, 30)
colors = ("Red", "Green", "Blue")
single = (5,)  # Must have comma for single item

print(coordinates)
print(type(coordinates))  # <class 'tuple'>

# Accessing items (same as list)
print(coordinates[0])   # 10
print(colors[-1])       # Blue

# Cannot change items
# colors[0] = "Yellow"  # This will give ERROR!

# Length
print(len(coordinates))  # 3

# Why use tuples?
# - Faster than lists
# - Can use as dictionary keys
# - Protect data from accidental changes
```

## 7 DICTIONARY (Key-Value Pairs)



```python
# Dictionary stores data as key-value pairs
# Like a real dictionary with words and meanings

person = {
    "name": "Nirmal",
    "age": 18,
    "city": "Kathmandu",
    "is_student": True
}

print(person)
print(type(person))  # <class 'dict'>

# Accessing values using keys
print(person["name"])      # Nirmal
print(person["age"])       # 18
print(person.get("city"))  # Kathmandu
print(person.get("country", "Nepal"))  # Nepal (default value)

# Adding new items
person["phone"] = 9762297240
print(person)

# Changing values
person["age"] = 19
print(person)

# Removing items
del person["city"]
print(person)

person.pop("phone")
print(person)

# Getting all keys
print(person.keys())       # dict_keys(['name', 'age', 'is_student'])

# Getting all values
print(person.values())     # dict_values(['Raj', 26, True])

# Getting key-value pairs
print(person.items())      # dict_items([('name', 'Raj'), ...])

# Loop through dictionary
for key, value in person.items():
    print(f"{key}: {value}")

# Another way
for key in person:
    print(key, person[key])
```

## 8 SET (Unique Collection)

```python
# Set contains unique values (no duplicates)
# Order is not guaranteed

numbers = {1, 2, 3, 4, 5}
fruits = {"Apple", "Banana", "Orange", "Apple"}  # Apple duplicate

print(numbers)
print(fruits)  # {"Apple", "Banana", "Orange"} (Apple appears once)
print(type(fruits))  # <class 'set'>

# Adding items
numbers.add(6)
print(numbers)

# Removing items
numbers.remove(1)  # Error if not found
numbers.discard(1)  # No error if not found

# Set operations (like in math)
set1 = {1, 2, 3}
set2 = {2, 3, 4}

print(set1 & set2)   # {2, 3} (intersection - common)
print(set1 | set2)   # {1, 2, 3, 4} (union - all)
print(set1 - set2)   # {1} (difference - only in set1)
print(set1 ^ set2)   # {1, 4} (symmetric difference)

# Checking membership
print(1 in set1)     # True
print(5 in set1)     # False
```

## TYPE CASTING

-  TYPE CASTING - Converting Data Types

        What is Type Casting?

          Type casting means converting one data type to another. For example, converting the number 10 (integer) to "10" (string).

## 1) INTEGER to STRING

```python
# Convert integer to string using str()

age = 25
age_string = str(age)

print(age)
print(type(age))           # <class 'int'>

print(age_string)
print(type(age_string))    # <class 'str'>

# Now you can concatenate with text
message = "My age is " + age_string + " years"
print(message)

# Or use f-string (automatic conversion)
message2 = f"My age is {age} years"
print(message2)
```

## 2) STRING to INTEGER



```python
# Convert string to integer using int()

number_string = "42"
number = int(number_string)

print(number_string)
print(type(number_string))  # <class 'str'>

print(number)
print(type(number))         # <class 'int'>

# Now you can do math
result = number + 8
print(result)  # 50

# CAUTION: String must contain only digits
# This will ERROR:
# bad = int("hello")

# For decimal strings, use float first
decimal_string = "10.5"
decimal = float(decimal_string)
integer = int(decimal)
print(integer)  # 10 (decimal part removed)
```

## 3) STRING to FLOAT

```python
# Convert string to float using float()

price_string = "99.99"
price = float(price_string)

print(price_string)
print(type(price_string))  # <class 'str'>

print(price)
print(type(price))         # <class 'float'>

# Math with float
total = price + 0.01
print(total)  # 100.0
```

## 4) FLOAT to INTEGER



```python
# Convert float to integer using int()
# Note: Decimal part is removed (not rounded)

height = 5.9
height_int = int(height)

print(height)
print(type(height))        # <class 'float'>
s
print(height_int)
print(type(height_int))    # <class 'int'>
print(height_int)          # 5 (decimal part lost)

# To round instead
rounded = round(height)
print(rounded)             # 6 (nearest whole number)

# Round to specific decimal places
price = 99.9999
print(round(price, 2))     # 100.0 (rounded to 2 decimals)
```

## 5) BOOLEAN CASTING




```python
# Convert boolean to other types

is_active = True
converted = int(is_active)

print(is_active)
print(converted)           # 1 (True = 1)

is_inactive = False
converted2 = int(is_inactive)
print(converted2)          # 0 (False = 0)

# String to boolean
text = "Hello"
print(bool(text))          # True (non-empty = True)

empty_text = ""
print(bool(empty_text))    # False (empty = False)

# Number to boolean
print(bool(10))            # True
print(bool(0))             # False (only 0 = False)
print(bool(-5))            # True
print(bool(0.0))           # False
```

## 6) COLLECTION CASTING

```python
# Convert between list, tuple, set

# List to Tuple
my_list = [1, 2, 3, 4, 5]
my_tuple = tuple(my_list)
print(my_tuple)            # (1, 2, 3, 4, 5)
print(type(my_tuple))      # <class 'tuple'>

# Tuple to List
my_tuple2 = (10, 20, 30)
my_list2 = list(my_tuple2)
print(my_list2)            # [10, 20, 30]

# List to Set (removes duplicates)
numbers_list = [1, 2, 2, 3, 3, 3, 4]
numbers_set = set(numbers_list)
print(numbers_set)         # {1, 2, 3, 4}

# String to List (each character becomes item)
text = "Hello"
text_list = list(text)
print(text_list)           # ['H', 'e', 'l', 'l', 'o']

# List to String
words = ["Hello", "World"]
sentence = " ".join(words)
print(sentence)            # "Hello World"
```

### ➕ PYTHON OPERATORS

# What are Operators?

    - Operators are symbols that perform operations. Like +, -, *, / etc.

## 1) ARITHMETIC OPERATORS (Math Operations)



```python
# Basic math operations

a = 20
b = 3

# Addition
print(a + b)   # 23

# Subtraction
print(a - b)   # 17

# Multiplication
print(a * b)   # 60

# Division (always returns float)
print(a / b)   # 6.666...

# Floor Division (returns integer)
print(a // b)  # 6 (removes decimal)

# Modulus (remainder)
print(a % b)   # 2 (20 divided by 3 leaves remainder 2)

# Exponentiation (power)
print(a ** b)  # 8000 (20 to the power 3)

# Operator precedence (order)
result = 10 + 5 * 2
print(result)  # 20 (multiply first, then add)

result2 = (10 + 5) * 2
print(result2)  # 30 (brackets first)

# More examples
print(2 + 3 * 4)      # 14 (multiply first: 3*4=12, then 2+12)
print(10 / 2 - 3)     # 2.0 (divide first: 10/2=5, then 5-3)
print(2 ** 3 + 1)     # 9 (power first: 2**3=8, then 8+1)
```

## 2) COMPARISON OPERATORS (Comparing Values)




```python
# These return True or False

a = 20
b = 30

# Equal to
print(a == b)  # False

# Not equal to
print(a != b)  # True

# Greater than
print(a > b)   # False

# Less than
print(a < b)   # True

# Greater than or equal
print(a >= b)  # False

# Less than or equal
print(a <= b)  # True

# Real-life examples
age = 18
print(age >= 18)  # True (adult check)

password = "hello123"
correct = "hello123"
print(password == correct)  # True

score = 85
print(score > 80)  # True
print(score <= 100)  # True
```

## 3) ASSIGNMENT OPERATORS (Assign Values)

```python
# Simple assignment
x = 10
print(x)  # 10

# Add and assign (+=)
x = 10
x += 5  # Same as x = x + 5
print(x)  # 15

# Subtract and assign (-=)
x = 10
x -= 3  # Same as x = x - 3
print(x)  # 7

# Multiply and assign (*=)
x = 10
x *= 2  # Same as x = x * 2
print(x)  # 20

# Divide and assign (/=)
x = 20
x /= 4  # Same as x = x / 4
print(x)  # 5.0

# Floor divide and assign (//=)
x = 20
x //= 3  # Same as x = x // 3
print(x)  # 6

# Modulus and assign (%=)
x = 20
x %= 3  # Same as x = x % 3
print(x)  # 2

# Power and assign (**=)
x = 2
x **= 3  # Same as x = x ** 3
print(x)  # 8

# Real example - counter
count = 0
count += 1  # count = 1
count += 1  # count = 2
count += 1  # count = 3
print(count)  # 3

# Score calculation
score = 0
score += 10  # Got 10 points
score += 5   # Got 5 more
score -= 2   # Lost 2 points
print(score)  # 13
```

## 4) IDENTITY OPERATORS (Is it the Same Object?)

```python
# Check if two variables point to same object

a = [1, 2, 3]
b = [1, 2, 3]
c = a

# is - same object (same memory location)
print(a is b)  # False (different lists)
print(a is c)  # True (c refers to same list as a)

# is not - different objects
print(a is not b)  # True

# None check
x = None
print(x is None)  # True

y = 5
print(y is None)  # False

# Important note
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True (same content)
print(a is b)  # False (different objects)
```

## 6) MEMBERSHIP OPERATORS (Is it in the Collection?)

```python
# Check if item exists in list/string/dict

# in - item exists
fruits = ["Apple", "Banana", "Orange"]
print("Apple" in fruits)     # True
print("Mango" in fruits)     # False

text = "Hello"
print("H" in text)           # True
print("X" in text)           # False

numbers = [10, 20, 30, 40]
print(25 in numbers)         # False
print(30 in numbers)         # True

# Dictionary check (checks keys by default)
person = {"name": "Raj", "age": 25}
print("name" in person)      # True
print("Raj" in person)       # False (checks keys, not values)

# not in - item doesn't exist
print("Mango" not in fruits)  # True
print("X" not in text)        # True
```
